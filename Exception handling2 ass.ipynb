Q1- Explain why we have to use the Exception class while creating a Custom Exception.

Custom exceptions enhance the usability of a class. An exception class should have init and str magic methods which are called automatically during the exception handling process. It is completely up to us what we want our custom error class to do,
we have the ability to create our own exception classes. The construction of custom exception classes can enrich our class design. A custom error class could logs errors, inspect an object. It is up to us what the exception class does, although a custom class will not usually do a great deal more than display a message.

In Python, we can define custom exceptions by creating a new class that is derived from the built-in Exception class.


```python
#for example 

#Exception is a base class which is inherited by LoadError class
class LoadError(Exception):
    """General exception to be used when there is an error with truck load."""
#OverloadTruckError class is inherited from the LoadError class
class OverloadTruckError(LoadError):
    """The truck is overloaded."""
#OverloadTruckError class is inherited from the LoadError class
class NoLoadOnTruckError(LoadError):
    """The truck is empty."""
```

```python
#class OverloadTruckError
load = 90
try:
    
    if load >50:
        raise OverloadTruckError
except OverloadTruckError:
    print(" Error: there is high load on truck")

# NoLoadOnTruckError class 
load = 0
try:
    
    if load == 0:
        raise NoLoadOnTruckError
except NoLoadOnTruckError:
    print(" Error: there is no  load on truck")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

You can use LookupError exception class to handle both IndexError and KeyError exception classes.

- LookupError

   --> IndexError
 
   --> KeyError

```python
#Index Error
try:
    list2 = [23,45.4,"shi",True]
    list2[7]
except LookupError as e:
    print("index not found",e.__class__)
```

```python
#Key Error

try:
    k = {1:"shi",2:"nitin",3:"anmol",4:"bittu"}
    print (k[5])
except KeyError as e:
    print ("key not found", e.__class__)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

#ImportError

In Python, ImportError occurs when the Python program tries to import module which does not exist in the private table. This exception can be avoided using exception handling using try and except blocks. We also saw examples of how the ImportError occurs and how it is handled.

```python
try:
    import shivangi
except ImportError as e:
    print ("shivangi module does'nt exist",e.__class__)
```

#ModuleNotFoundError

The 'module not found' error is a syntax error that appears when the static import statement cannot find the file at the declared path.

The 'module not found' error in Python occurs when the interpreter is unable to import the specified module.

Q6. List down some best practices for exception handling in python.

```python
# 1 - use always a specific exception
try :
    10/0
except ZeroDivisionError as e :#use always a specific exception in place of Exception
    print(e)
```

```python
#2 - always try to log your error
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR) # use log in place of print
try :
    10/0
except ZeroDivisionError as e :
    logging.error("i am trying to handle a zerodivision error {} ".format(e) )
```

```python
#3 - alwyas avoid to write a multiple exception handling 
try :
    10/0
# use only that error which is really going to use in the case
except FileNotFoundError as e : 
    logging.error("i am handling file not found  {} ".format(e) )
except AttributeError as e : 
    logging.error("i am handling Attribute erro  {} ".format(e) )
except ZeroDivisionError as e :
    logging.error("i am trying to handle a zerodivision error {} ".format(e) )
```
